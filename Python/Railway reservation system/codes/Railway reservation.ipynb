```python
class Train:
    def __init__(self, train_number, train_name, seats_available):
        self.train_number = train_number
        self.train_name = train_name
        self.seats_available = seats_available
    
    def check_availability(self, requested_seats):
        if requested_seats <= self.seats_available:
            self.seats_available -= requested_seats
            return True
        else:
            return False

# Define list of trains
trains = [Train(1, 'Rajdhani Express', 200),Train(2, 'Shatabdi Express', 150),
          Train(3, 'Duronto Express', 100)]

# Function to show available trains
def show_available_trains():
    for index, train in enumerate(trains):
        print(f'{index+1}. Train Number: {train.train_number}, Train Name: {train.train_name}, 
              Seats Available: {train.seats_available}')

# Function to make reservation
def make_reservation():
    while True:
        show_available_trains()
        print("Enter 0 For Exit from Loop.")
        print("\n------------------------------------------------------------")
        train_choice = int(input("Enter the number of the train you want to reserve: "))
        if train_choice == 0:
            break

        selected_train = trains[train_choice - 1]
        requested_seats = int(input("Enter the number of seats you want to reserve: "))
        if selected_train.check_availability(requested_seats):
            print(f'{requested_seats} seats reserved in {selected_train.train_name} 
                  ({selected_train.train_number})')
            print("------------------------------------------------------------\n")

        else:
            print(f'Sorry, only {selected_train.seats_available} seats available in 
                  {selected_train.train_name} ({selected_train.train_number})')
```

```python
make_reservation()
show_available_trains()
```
